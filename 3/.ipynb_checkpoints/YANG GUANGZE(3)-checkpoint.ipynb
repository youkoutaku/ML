```python
from scipy.stats import t 
x_bar = 157.77 
alpha = 0.05 
s2 = 26.419 
N = 60 
m = N-1 # 自由度
t = t.ppf( 1-alpha/2, df=m ) # カウントの向きが逆なので
Left = x_bar - t*np.sqrt(s2/N) 
Right = x_bar + t*np.sqrt(s2/N) 
[Left, Right]
```

```python
from scipy.stats import chi2 
x = [4.9, -2.6, 3.1, -1.2, 6.4, -3.5]
N = np.size(x) 
s2 = np.var(x, ddof=1) # 標本不偏分散
m = N-1 # 自由度 
alpha = 0.05 
chi2_L = chi2.ppf( 1-alpha/2, df=m ) # カウントの向きが逆なので
chi2_R = chi2.ppf( alpha/2, df=m ) # カウントの向きが逆なので
Left = (N-1)*s2/chi2_L 
Right = (N-1)*s2/chi2_R 
[Left, Right] 
```

```python
import numpy as np
from scipy.stats import t
x=[7.86, 7.89, 7.84, 7.90, 7.82]
x_bar = np.mean(x)
alpha = 0.01
s2 =np.var(x,ddof=1)
N = np.size(x)
m = N-1 # 自由度
t = t.ppf( 1-alpha/2, df=m ) # カウントの向きが逆なので
Left = x_bar - t*np.sqrt(s2/N)
Right = x_bar + t*np.sqrt(s2/N)
[Left, Right]
```

```python
import numpy as np
from scipy.stats import chi2
x=[15.4, 16.1, 15.7, 16.6, 14.9, 15.5, 16.2]
alpha = 0.01
s2 =np.var(x,ddof=1)
N = np.size(x)
m = N-1 # 自由度
chi2_L = chi2.ppf( 1-alpha/2, df=m ) # カウントの向きが逆なので
chi2_R = chi2.ppf( alpha/2, df=m )
Left = (N-1)*s2/chi2_L
Right = (N-1)*s2/chi2_R
[Left, Right]
```
